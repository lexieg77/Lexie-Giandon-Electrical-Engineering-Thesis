```python
import pandas as pd
import re

# Load CSVs
original_path = "/content/drive/My Drive/Colab Notebooks/f1_scores/original_500.csv"
original = pd.read_csv(original_path, encoding='latin1')

predicted_path = "/content/drive/My Drive/Colab Notebooks/f1_scores/new_500.csv"
predicted = pd.read_csv(predicted_path, encoding='latin1')

# Assume both have a column called 'part_name'
original_parts = set(original["Critical Spares"].astype(str).str.lower().str.strip())
predicted_parts = set(predicted["Critical Spares"].astype(str).str.lower().str.strip())

print(original_parts)
print(predicted_parts)

#cleaned_original = {re.findall(r'\d+', item)[-1] for item in original_parts if re.findall(r'\d+', item)} # only keep the part number
#cleaned_predicted = {item.replace("?", "") for item in predicted_parts} # remove ? between the part number

# Compute overlaps
true_positives = len(original_parts & predicted_parts)
false_positives = len(predicted_parts - original_parts)
false_negatives = len(original_parts - predicted_parts)


print(true_positives)
print(false_positives)
print(false_negatives)

precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1 = 2 * (precision * recall) / (precision + recall)

print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 Score: {f1:.3f}")
```
